```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
reliant_df= pd.read_csv("Resources/ReliantData.csv")
reliant_df
```

```python
Lead_df= pd.read_csv("Resources/LeadData.csv")
Lead_df
```

```python
Reliant_merge= pd.merge(reliant_df, Lead_df,how="left", on=["Lead ID", "Lead ID"])
Reliant_merge
```

```python
Reliant_merge.info()
```

```python
Reliant_merge.shape
```

```python
cost_lead_activated= Reliant_merge[Reliant_merge["Lead Status"]=='Activated']['Cost Per Lead'].sum()
cost_lead_activated
```

```python
cost_lead_confirmed= Reliant_merge[Reliant_merge["Lead Status"]=='Confirmed']['Cost Per Lead'].sum()
cost_lead_confirmed
```

```python
cost_lead_approved= Reliant_merge[Reliant_merge["Conversion Status"].isin(['Lost','Funded'])]['Cost Per Lead'].sum()
cost_lead_approved
```

```python
cost_lead_funded= Reliant_merge[Reliant_merge["Conversion Status"]=='Funded']['Cost Per Lead'].sum()
cost_lead_funded
```

```python
cost_lead= Reliant_merge['Cost Per Lead'].sum()
cost_lead
```

```python
cost_lead_credit= Reliant_merge.groupby(["SelfGradedCredit"]).sum()['Cost Per Lead']
cost_lead_credit
```

```python
cost_lead_time_less24= Reliant_merge[Reliant_merge["Time in Business (Months)"] < 24].sum()['Cost Per Lead']
cost_lead_time_less24
```

```python
cost_lead_time_greater24= Reliant_merge[Reliant_merge["Time in Business (Months)"] >= 24].sum()['Cost Per Lead']
cost_lead_time_greater24
```

```python
SelfGradedCredit_dict = {"Excellent":0, "Good":1, "Fair":2, "Poor":3}
Reliant_merge['SelfGradedCredit'].replace(SelfGradedCredit_dict, inplace=True)

Reliant_merge.head(10)
```

```python
BusinessType_dict = {"A Corp":0 , "B Corp":1 , "C Corp": 2 , "D Corp": 3 , "E Corp": 4}
Reliant_merge['Business Type'].replace(BusinessType_dict, inplace=True)

Reliant_merge.head(10)
```

```python
Reliant_merge.to_csv('Reliant_merge.csv',index=False)
```
